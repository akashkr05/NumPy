# Random sampling with numpy

```python
import numpy as np
import random
```

```python
np.random.random(1)
```

```python
np.random.random(3)
```

```python
np.random.random(3).ndim
```

```python
np.random.random(1).ndim
```

```python
np.random.random((3,3))
```

```python
np.random.random((3,3)).ndim
```

```python
np.random.randint(1)
```

```python
np.random.randint(1, 4, (10, 10)) # 1,4 is range and (10,10) is shape
```

```python
np.random.randint(1,4,(2,4,4))  # Create 3D array
```

```python
np.random.seed()   # seed function access always same value as above if we use same range and shape
np.random.randint(1,4,(2,4,4))
```

```python
np.random.seed()  
np.random.randint(1,4,(2,4,4))
```

```python
np.random.rand(5) # rand() function gives 0 to 1 value
```

```python
np.random.rand(4,4) 
```

```python
x = [1, 2, 3, 4, 5]
np.random.choice(x)
```

```python
x = [1, 2, 4, 6, 5, 7, 10, 9, 13]
np.random.choice(x)
```

```python
np.random.seed(60)
for i in range(20):
    print(np.random.choice(x))
```

```python
np.random.randn(4, 4)  # randn() function gives both the positive and negative value
```

```python
x = [ 7, 10, 13,  1,  5,  2,  4,  9,  6]
np.random.permutation(x)
```

```python
np.random.seed(60)
for i in range(20):
    print(np.random.choice(x))
```

```python
np.random.seed(60)
for i in range(44):
    print(np.random.choice(x))
```
