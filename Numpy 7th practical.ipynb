# PLOT GRAPH

# sin plot

```python
x_sin = np.arange(0, 4*np.pi, 0.4)
print(x_sin)
```

```python
print(type(x_sin))
```

```python
y_sin = np.sin(x_sin)
print(y_sin)
```

```python
plt.plot(x_sin, y_sin)
plt.show()
```

# Cos plot

```python
y_cos = np.cos(x_sin)
print(y_cos)
```

```python
plt.plot(x_sin, y_cos)
plt.show()
```

# Tan plot

```python
y_tan = np.tan(x_sin)
print(y_tan)
```

```python
plt.plot(x_sin, y_tan)
plt.show()
```
